```python
%%writefile Lec1DataTypes.scala

object DataType{
    def main(args: Array[String]): Unit={
        println("Hello World")

        var a: Int = 10
        var b:Float = 20
        var c: String="Hello World"
        var d: Byte=10
        val e: Boolean=true

        println("The data type of a is: ", a.getClass.getName)
        println("The data type of b is: ", b.getClass.getName)
        println("The data type of c is: ", c.getClass.getName)
        println("The data type of d is: ", d.getClass.getName)
        println("The data type of e is: ", e.getClass.getName)
    }
}
```

```python
!scalac /content/Lec1DataTypes.scala
```

```python
!scala DataType
```

```python
from pyspark import SparkContext, SparkConf

conf = SparkConf().setAppName("Demo")
sc = SparkContext.getOrCreate(conf=conf)
rdd = sc.textFile("/content/Lec1DataTypes.scala")
rdd.collect()
```

```python
rdd1 = rdd.filter(lambda x: x!="")
rdd1.collect()
```
